# วิเคราะห์การถดถอยเชิงเส้น (Linear Regression)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# สร้างอาร์เรย์ที่มีค่าเป็นลำดับของตัวเลขตั้งแต่ -5 ถึง 5 โดยมีจำนวน 100 ค่า
x = np.linspace(-5, 5, 100)
```

```python
# y=ax+b

a = 2
b = 1
y = a * x + b

plt.plot(x, y, label='y = 2x + 1', color='blue')
plt.title('Linear Function Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
rng = np.random
x = rng.rand(50) * 10
a = rng.rand(50)
b = rng.rand(50) 
y = a * x + b

# plt.scatter(x, y, label='y = 2x + 1', color='blue')
plt.scatter(x, y)
plt.title('Scatter Plot of Linear Function')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
